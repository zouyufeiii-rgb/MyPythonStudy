# Dictionaries

```python
licensePlates = ["AN","BO","CH","DA"]
colors = ["white","black","white","white"]
zip(licensePlates,colors) #zip生成的是 tuple（元组）
list(zip(licensePlates,colors))
```

```python
dict(zip(licensePlates,colors))
```

```python
# licensePlate2Color
m = dict(zip(licensePlates,colors))
```

```python
m["AN"]
```

```python
len(m) #4个键值对
```

```python
m["BO"]
```

```python
m["XX"]
```

```python
m.get("AN","uknown")
#“去字典 m 中寻找键（Key）为 "AN" 的值，如果没找到，就给我返回 "unknown"”
```

```python
m.keys()
```

```python
list(m.keys())
```

```python
m.values()
```

```python
list(m.values())
```
